# Fundamental concept 

For a stochastic process $Y_t$, define:

1. **Mean**: $$ \mu_t=EY_t $$
2. **autocovariaance**: $$ \gamma _ { t , s } = \operatorname { Cov } \left( Y _ { t } , Y _ { s } \right) $$
3. **autocorrelation**: $$ \rho _ { t , s } = \operatorname { Corr } \left( Y _ { t } , Y _ { s } \right)=\frac { \operatorname { Cov } \left( Y _ { t } , Y _ { s } \right) } { \sqrt { \operatorname { Var } \left( Y _ { t } \right) \operatorname { Var } \left( Y _ { s } \right) } } = \frac { \gamma _ { t , s } } { \sqrt { \gamma _ { t , t } \gamma _ { s , s } } } $$

## Random walk
Let $e_{1:}$ be a sequnce of i.i.d. r.v.s' with zero mean and $\sigma^2$ variance. Then constructed time series $Y_t$ as: $$ Y_t=\sum_{i=1}^t e_i $$
Hence we have:
$$ \mu_t=E\sum e_i=\sum Ee_i =0 $$ $$ \text{Var} (Y_t)= \operatorname{Var}(\sum e_i)=\sum \operatorname{Var}e_i=t\sigma^2\\ \gamma_{t,s}=\sum _ { i = 1 } ^ { S } \sum _ { j = 1 } ^ { t } C o v \left( e _ { i } , e _ { j } \right)=\min(t\sigma^2, s\sigma^2)\\ \rho_{t,s}=\min(\sqrt{\frac{t}{s}},\sqrt{\frac{s}{t}})$$

## Stationarity

**Strietly stationary:** For all $k\ge 0, n\ge0$, subsequence of timeseries $Y_{t_1:}=\{Y_{t1},Y_{t2},\cdots Y_{t_n}\}$ and $\{Y_{t1-k},Y_{t2-k},\cdots Y_{t_n-k}\}$ has the same joint distribution.

**Weakly(second-order) stationary**: Forall $k\ge 0$: $$ \mu_t=\mu_{t\pm k}\\ \gamma_{t,t-k}=\gamma_{0,k} $$

## White Noise

White noise process is a sequence of i.i.d. r.v.s', which is strictly stationary clearly. And
$$ \gamma _ { k } = \left\{ \begin{array} { c c } \operatorname { Var } \left( e _ { t } \right) & \text { for } k = 0 \\ 0 & \text { for } k \neq 0 \end{array} \right. $$ We usually take $\mu_t=0$.

## Random Cosine Wave
The following time series example suggests that it isn't reliable to judge satationarity from plots. Suppose process: $$ Y _ { t } = \cos \left[ 2 \pi \left( \frac { t } { 12 } + \Phi \right) \right] \quad \text { for } t = 0 , \pm 1 , \pm 2 $$ where $\Phi \sim U(0,1)$ and it is selected before the process. We have $$ E \left( Y _ { t } \right) = E \left\{ \cos \left[ 2 \pi \left( \frac { t } { 12 } + \Phi \right) \right] \right\}= \int_0^1 \cos \left[ 2 \pi \left( \frac { t } { 12 } + \phi \right) \right] d \phi =0\\\gamma _ { t , s } = E \left\{ \cos \left[ 2 \pi \left( \frac { t } { 12 } + \Phi \right) \right] \cos \left[ 2 \pi \left( \frac { s } { 12 } + \Phi \right) \right] \right\}=\frac { 1 } { 2 } \cos \left[ 2 \pi \left( \frac { | t - s | } { 12 } \right) \right]$$ Hence this process is weakly stationary.
