```r
require(Hmisc)
require(repr)
require(stats)
require(ggplot2)
require(ggpubr)
require(tidyverse)
require(lubridate)
```

```r
larain <- read.table("../data/larain.dat", header=TRUE)
```

```r
larain <- larain %>%
mutate(date=1878:(1878+114))
```

```r
options(repr.plot.height=4)
ggplot(data=larain)+
    geom_line(aes(x=date,y=larain))+
    xlab("year")+ylab("inch")
```

```r
# options(repr.plot.width=9, repr.plot.height=9)
# plot(y=larain, x=Lag(larain), ylab='Inches', xlab='Previous Year Inches', col='blue')

ggplot(data=larain) + 
  geom_point(aes(x=Lag(larain), y=larain), shape=1,na.rm=T) +
  xlab('Previous Year Inches') + ylab('Inches')+
  coord_equal()
```

```r
color <- read.table("../data/color.dat", header=TRUE)
```

```r
len <- nrow(color)
color <- color %>%
    mutate(batch=1:len)
```

```r
# options(repr.plot.width=10, repr.plot.height=3)
ggplot(data=color,map=aes(x=batch,y=color))+
    geom_point(shape=1)+
    geom_line()
#     coord_equal()
```

```r
ggplot(data=color,map=aes(x=lag(color),y=color))+
    geom_point(shape=1,na.rm=T)+
    coord_fixed()+
    xlab("prev batch color")
#     geom_line()
```

## Hare

```r
hare <- read.table("../data/hare.dat", header=TRUE)
```

```r
years <- nrow(hare)
hare <- hare %>%
    mutate(year=1905:(1905+years-1))
```

```r
# options(repr.plot.height=4)
ggplot(data=hare,map=aes(x=year,y=hare))+
    geom_line()+
    geom_point(shape=1)+
    xlab("year")+ylab("Abundance")+
    scale_y_continuous(breaks=seq(0, max(hare), 20))
```

```r
ggplot(data=hare,map=aes(x=lag(hare),y=hare))+
    geom_point(shape=1,na.rm=T)+
    coord_fixed()+
    xlab("prev year abundance")
#     geom_line()
```

## Tempdub

```r
tempdub <- read.table("../data/tempdub.dat", header=TRUE)
```

```r
tempdub <- tempdub %>% 
    mutate(date=seq(ymd(19640101),ymd(19751201),by="month"))
```

```r
options(repr.plot.width=12,repr.plot.height=4)
ggplot(data=tempdub,map=aes(x=date,y=tempdub))+
    geom_line()+
    geom_point(shape=1)+
    xlab("year")+ylab("temp")+
    theme_pubclean()
#     scale_y_continuous(breaks=seq(0, max(tempdub), 20))
```
