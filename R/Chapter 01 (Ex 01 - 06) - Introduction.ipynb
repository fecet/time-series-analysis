**Exercise 1.1**.  Use software to produce the time series in Exhibit 1.2, on page 2.  The data are in the file named `larain`.

```r
require(Hmisc)
require(repr)
require(stats)
require(ggplot2)
require(ggpubr)
```

```r
larain <- read.table("../data/larain.dat", header=TRUE)[['larain']]
options(repr.plot.width=12, repr.plot.height=8)
# plot(y=larain, x=Lag(larain), ylab='Inches', xlab='Previous Year Inches', col='blue')

ggplot() + 
  geom_point(aes(x=Lag(larain), y=larain), color='blue', na.rm=T) +
  xlab('Previous Year Inches') + ylab('Inches') +
  theme_bw() + theme(text = element_text(size=16))
```

**Exercise 1.3**.  Simulate a completely random process of length 48 with independent, normal values.  Plot the time series plot.  Does it look "random"?  Repeat this exercise several times with a new simulation each time.

```r
n = 48
set.seed(0)

plots = list()
options(repr.plot.width=14, repr.plot.height=14)

X <- function() rnorm(n, mean=0, sd = 1)
for(i in 1:5) {
  p <- ggplot() + 
    geom_line(aes(x = 1:n, y = X()), color="blue") + 
    xlab('Observation') + ylab('Value') +
    theme_bw() + theme(text = element_text(size=16))
  plots[[i]] <- p
}

ggarrange(plotlist=plots, nrow=length(plots), ncol=1)
```

See the concept of [Apophenia](https://en.wikipedia.org/wiki/Apophenia) for a discussion on seeing patterns in random data.

**Exercise 1.4**.  Simulate a completely random process of length 48 with independent, chi-square distributed values, each with 2 degrees of freedom.  Display the time series plot.  Does it look "random" and nonnormal?  Repeat this exercise several times with a new simulation each time.

```r
n = 48
set.seed(0)

plots = list()
options(repr.plot.width=14, repr.plot.height=14)

X <- function() rchisq(n, df=2)
for(i in 1:5) {
  p <- ggplot() + 
    geom_line(aes(x = 1:n, y = X()), color="blue") + 
    xlab('Observation') + ylab('Value') +
    theme_bw() + theme(text = element_text(size=16))
  plots[[i]] <- p
}

ggarrange(plotlist=plots, nrow=length(plots), ncol=1)
```

**Exercise 1.5**.  Simulate a completely random process of length 48 with independent, t-distributed values, each with 5 degrees of freedom.  Display the time series plot.  Does it look "random" and nonnormal?  Repeat this exercise several times with a new simulation each time.

```r
n = 48
set.seed(0)

plots = list()
options(repr.plot.width=14, repr.plot.height=14)

X <- function() rt(n, df=5)
for(i in 1:5) {
  p <- ggplot() + 
    geom_line(aes(x = 1:n, y = X()), color="blue") + 
    xlab('Observation') + ylab('Value') +
    theme_bw() + theme(text = element_text(size=16))
  plots[[i]] <- p
}

ggarrange(plotlist=plots, nrow=length(plots), ncol=1)
```
